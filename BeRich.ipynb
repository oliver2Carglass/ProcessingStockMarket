```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Télécharger les données du S&P 500
sp500 = yf.download('^GSPC', start='2000-01-01', end='2024-12-31')

# Optionnel : sauvegarder en CSV
sp500.to_csv('SP500_data.csv')
sp500.columns = sp500.columns.droplevel(1)
sp500
```

```python
# Télécharger les données du NASDAQ Composite (^IXIC)
nasdaq = yf.download('^IXIC', start='2000-01-01', end='2024-12-31')



# Optionnel : sauvegarder les données en CSV
nasdaq.to_csv('nasdaq_composite.csv')
# Enlève le niveau 'Ticker' (niveau 1) du MultiIndex sur les colonnes
nasdaq.columns = nasdaq.columns.droplevel(1)

nasdaq
```

```python
def findVolatilityAndAmplitude(df):
    df["Amplitude"] = df["High"] - df["Low"]
    df["Volatility(%)"]=df["Amplitude"]/df["Open"]*100

```

```python
findVolatilityAndAmplitude(nasdaq)
findVolatilityAndAmplitude(sp500)
```

```python
%matplotlib inline
plt.figure(figsize=(15, 6))
# Trace la volatilité du NASDAQ
plt.plot(nasdaq.index, nasdaq['Volatility(%)'], label='NASDAQ Volatility', color='blue', alpha=0.7)

# Trace la volatilité du S&P 500
plt.plot(sp500.index, sp500['Volatility(%)'], label='S&P 500 Volatility', color='orange', alpha=0.7)


plt.title('Volatility Comparison: NASDAQ vs S&P 500')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.legend()
plt.grid(True)
plt.show()
```

```python
sp5M=sp500["Volatility(%)"].mean()
nsdM=nasdaq['Volatility(%)'].mean()
```
